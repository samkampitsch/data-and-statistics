```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt

# Load the data and filter for 'radius' and 'mass'
data = pd.read_csv(r'C:\Users\Anastasia Kiselev\Downloads\Tabellen\exoplanet.eu_catalog.csv')
filtered_data = data[(data['radius'] < 1) & (data['mass'] < 1)]

# Transform data and perform linear regression
x= np.log10(filtered_data['radius']) 
y= np.log10(filtered_data['mass'])
model = sm.OLS(y, sm.add_constant(x))
results= model.fit()
print(results.summary())
# Bootstrap method for confidence interval
num_bootstrap_samples = 200
bootstrap_slopes = []

for _ in range(num_bootstrap_samples):
    indices = np.random.choice(len(x), len(x), replace=True)
    x_bootstrap, y_bootstrap = x.iloc[indices], y.iloc[indices]
    results_bootstrap = sm.OLS(y_bootstrap, sm.add_constant(x_bootstrap)).fit()
    bootstrap_slopes.append(results_bootstrap.params.iloc[1])  # Updated line

# Calculate 1-sigma confidence interval for the slope
confidence_interval = np.percentile(bootstrap_slopes, [15.85, 84.15])

# Calculate Pearson correlation coefficient excluding non-numeric columns
numeric_data = filtered_data.select_dtypes(include=[np.number])
corr_coef = numeric_data.corr().iloc[0, 1]

# Print results
print(f"Slope: {results.params.iloc[1]:.3f} +/- {results.bse.iloc[1]:.3f}")  # Updated line
print(f"Intercept: {results.params.iloc[0]:.3f} +/- {results.bse.iloc[0]:.3f}")  # Updated line
print(f"Pearson Correlation Coefficient: {corr_coef:.3f}")
print(f"1-sigma Confidence Interval for the Slope: {confidence_interval}")

# Plotting
plt.scatter(x, y, label='Data')
plt.plot(x, results.params.iloc[0] + results.params.iloc[1] * x, color='black', label='Linear Regression')  # Updated line
plt.fill_between(x, results.params.iloc[0] + confidence_interval[0] * x, results.params.iloc[0] + confidence_interval[1] * x, color='red', alpha=0.2, label='Confidence Interval')
plt.xlabel('log(Radius)')
plt.ylabel('log(Mass)')
plt.legend()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data and filter for 'radius' and 'mass'
data = pd.read_csv(r'C:\Users\Anastasia Kiselev\Downloads\Tabellen\exoplanet.eu_catalog.csv')
filtered_data = data[(data['radius'] < 1) & (data['mass'] < 1)]
print(len(filtered_data))
# Transform data and perform linear regression
x = np.log10(filtered_data['radius'])
y = np.log10(filtered_data['mass'])
model = sm.OLS(y, sm.add_constant(x))
results = model.fit()
print(results.summary())

# Bootstrap method for confidence interval
num_bootstrap_samples = 100
bootstrap_slopes = []

for _ in range(num_bootstrap_samples):
    indices = np.random.choice(len(x), len(x), replace=True)
    x_bootstrap, y_bootstrap = x.iloc[indices], y.iloc[indices]
    results_bootstrap = sm.OLS(y_bootstrap, sm.add_constant(x_bootstrap)).fit()
    bootstrap_slopes.append(results_bootstrap.params.iloc[1])

# Calculate 1-sigma confidence interval for the slope
confidence_interval = np.percentile(bootstrap_slopes, [15.85, 84.15])

# Calculate Pearson correlation coefficient excluding non-numeric columns
numeric_data = filtered_data.select_dtypes(include=[np.number])
corr_coef = numeric_data.corr().iloc[0, 1]

# Calculate covariance of the data set
covariance = np.cov(x, y)[0, 1]

# Plotting
plt.figure(figsize=(10, 6))

# Gaussian-type plot for Bootstrap Slopes
sns.kdeplot(bootstrap_slopes, color='blue', fill=True, label='Bootstrap Slopes KDE')

# Bootstrap Mean
bootstrap_mean = np.mean(bootstrap_slopes)
plt.axvline(bootstrap_mean, color='green', linestyle='--', label='Bootstrap Mean')

# Confidence Interval
plt.axvline(confidence_interval[0], color='red', linestyle='--', label='Confidence Interval')
plt.axvline(confidence_interval[1], color='red', linestyle='--')
plt.axvline(1.5343)

plt.xlabel('Bootstrap Slopes')
plt.ylabel('Density')
plt.legend()
plt.title('Bootstrap Slopes Distribution')


# Print additional results
print(f"Slope: {results.params.iloc[1]:.3f} +/- {results.bse.iloc[1]:.3f}")
print(f"Y-Axis Offset: {results.params.iloc[0]:.3f} +/- {results.bse.iloc[0]:.3f}")
print(f"Covariance: {covariance:.3f}")
plt.show()

print(bootstrap_mean)
```
