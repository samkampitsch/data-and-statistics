```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from astropy.io import ascii, fits
```

```python

file_path = 'Radio Fluxes.dat'
data = pd.read_csv(file_path)

df = pd.read_csv(file_path, sep='\s+')
df

```

```python
sum_G = df['G'].sum()
cumsum_G = df['G'].cumsum()

print(sum_G)
print(cumsum_G)
```

```python
Sm = cumsum_G / 386

print(Sm)
```

```python
sum_R = df['R'].sum()
cumsum_R = df['R'].cumsum()

print(sum_R)
print(cumsum_R)
```

```python
Sn = cumsum_R / 290

print(Sn)
```

```python
D = Sn - Sm
print(D)

D_max = np.max(D)
print(D_max)
```
