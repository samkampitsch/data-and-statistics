```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd

# Load the data from the CSV file
full_dataset = pd.read_csv(r'C:\Users\Anastasia Kiselev\Downloads\Tabellen\exoplanet.eu_catalog.csv')

# Specify the columns you want to keep
subset_columns = ['radius', 'mass']

# Create a new DataFrame with only the selected columns
selected_data = full_dataset[subset_columns]

# Extract subsets from the DataFrame
subset1 = selected_data['mass']
subset2 = selected_data['radius']

# Hypothesis: Mean of subset1 is different from the mean of subset2
# Null Hypothesis (H0): Mean of subset1 = Mean of subset2
# Alternative Hypothesis (H1): Mean of subset1 != Mean of subset2

# Calculate means and standard deviations
mean_subset1 = np.mean(subset1)
std_dev_subset1 = np.std(subset1)

mean_subset2 = np.mean(subset2)
std_dev_subset2 = np.std(subset2)

# Perform t-test
t_statistic, p_value = stats.ttest_ind(subset1, subset2, equal_var=False)  # Assuming unequal variances

# Display means and standard deviations
print(f"Mean of subset1 (mass): {mean_subset1}, Standard deviation of subset1 (mass): {std_dev_subset1}")
print(f"Mean of subset2 (radius): {mean_subset2}, Standard deviation of subset2 (radius): {std_dev_subset2}")

# Display t-statistic and p-value
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Plotting the data sets
plt.hist(subset1, bins=10, alpha=0.5, color='blue', label='Mass')
plt.hist(subset2, bins=10, alpha=0.5, color='orange', label='Radius')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Mass and Radius Subsets')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd

# Load your data or generate sample data
# For demonstration purposes, let's generate some random data
np.random.seed(42)
x = np.random.rand(100)
y = 3 * x + 2 + 0.1 * np.random.randn(100)  # True slope is 3, intercept is 2

# Hypothesized slope for the hypothesis test
hypothesized_slope = 3.2

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Hypothesis Test
# Null Hypothesis (H0): The slope is equal to the hypothesized_slope
# Alternative Hypothesis (H1): The slope is not equal to the hypothesized_slope

# Calculate the t-statistic for the slope
t_statistic_slope = (slope - hypothesized_slope) / std_err

# Degrees of freedom
df = len(x) - 2  # degrees of freedom for a simple linear regression

# Two-tailed p-value for the t-statistic
p_value_slope = 2 * (1 - stats.t.cdf(np.abs(t_statistic_slope), df=df))

# Display results
print(f"Observed slope: {slope}")
print(f"Standard Error of the slope: {std_err}")
print(f"T-statistic for the slope: {t_statistic_slope}")
print(f"P-value for the slope: {p_value_slope}")

# Plotting the data and the regression line
plt.scatter(x, y, label='Data')
plt.plot(x, intercept + slope * x, color='red', label='Linear Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression and Hypothesis Test')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd

# Load the data from the CSV file
full_dataset = pd.read_csv(r'C:\Users\Anastasia Kiselev\Downloads\Tabellen\exoplanet.eu_catalog.csv')

# Filter data for 'radius' > 1 and 'mass' > 1
filtered_data = full_dataset[(full_dataset['radius'] > 1) & (full_dataset['mass'] > 1)]

# Combine 'radius' and 'mass' datasets
combined_data = pd.concat([filtered_data['radius'], filtered_data['mass']], axis=1)

# Range of hypothesized slope values to test
hypothesized_slope_values = np.linspace(0.1, 1.0, 10)  # Adjust the range accordingly


# Lists to store results
observed_slopes = []
t_statistics = []
p_values = []

# Perform linear regression and hypothesis test for each hypothesized slope
for hypothesized_slope in hypothesized_slope_values:
    slope, intercept, _, _, std_err = stats.linregress(filtered_data['radius'], filtered_data['mass'])
    
    # Calculate t-statistic for the slope
    t_statistic = (slope - hypothesized_slope) / std_err
    
    # Degrees of freedom
    df = len(filtered_data) - 2
    
    # Two-tailed p-value for the t-statistic
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    # Store results
    observed_slopes.append(slope)
    t_statistics.append(t_statistic)
    p_values.append(p_value)

# Plot the results
plt.plot(hypothesized_slope_values, p_values, marker='o')
plt.axhline(0.05, color='red', linestyle='--', label='Significance Level (0.05)')
plt.xlabel('Hypothesized Slope')
plt.ylabel('P-value')
plt.title('Impact of Hypothesized Slope on Hypothesis Test')
plt.legend()
plt.show()
```
