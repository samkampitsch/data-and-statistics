### 1.1 Definition of Exoplanets 
To give a short definition of an exoplanet one would say it is a planet outside the solar system. Hence the also popular name "extrasolar planet" (wow so kreativ). Although the IAU has never officially agreed on a standard definition there are some statements they agreed on. 
1. The minimum size and mass are defined equally in exoplanets as in the solar system: ability to clear their orbit and the overcoming of external forces by sufficient self-gravity. 
2. Masses must lay below the minimum mass for thermonuclear fusion of deuterium (around 13 Jupiter masses). Anything above this limit is called a brown dwarf.
3. These objects must orbit stars or stellar remnants. Bodies with masses < 13M<sub>J</sub> that are free-floating in young star clusters are not considered to be planets but sub-brown dwarfs.

### 1.2 Past

The first exoplanet to be discovered by using the radial velocity method is 51 Pegasi b. This exoplanet was discovered in 1995 by Mayor and Queloz (abolute Gigachads). <br> 
In this method, the mass of the exoplanet and its influence on the orbital motion of the host star is being analyzed in the context of its spectra. Due to the additional mass of the exoplanet, the host star’s orbital motion around its mass center changes. Now, because of this, there are some changes visible in the spectra of the host star. Those changes are, for example, periodic blue or redshift. This shift in a spectrum in combination with movement is called the Doppler effect (oda ums in Sheldon Cooper’s Wörtern zu sogn: niuuuu). <br>
One of the downsides of this detection method is the detection of the exoplanet’s mass. Since the detected mass is always the minimum mass of the exoplanet, and its real mass can be way bigger than previously detected mass. Due to that, the changes in the orbital motions of the host star around its mass center are also not exactly detectable, and neither are its spectra changes. Another downside of the radial velocity method is the need for a very high-resolution spectrograph since changes in the spectra are often hardly detectible due to the changes being relatively small (dadat ma stinkn). [(Chien-Hsiu Lee 2018)](#ppf)

### 1.3 Present

Present experiments and missions concerning the detection of exoplanets and their systems are being focused on the precision of the radii measurements. This is achieved with the combination of the radial velocity method and the transit method. Because of this combination, the radius of the host star can be detected more precisely, and with that, the exoplanet’s radius gets detected more precisely as well (win-win situation my friend). The combination of those two detection methods also solves the problem of the exoplanet’s orbit inclination, which occurred in the radial velocity method. An example of such a mission is the California Kepler Survey (CKS), which uses the High-Resolution Echelle Spectrometer (HIRES).

Another important research is concerning the exoplanets’ orbit. It has been figured out that an exoplanet’s orbit can change over time (who would’ve thought).This can happen due to planetary scattering. Planetary scattering can only happen if at least two exoplanets orbit the same host star. If their orbits cross each other and following on from that collide. One of the exoplanets gets pushed away from the host star leading to an expanding orbit, and the other one gets pushed towards the host star leading to an increased orbit. [(Chien-Hsiu Lee 2018)](#ppf)<br>

### 1.5 First discovery of an exoplanet

In January 1992, a landmark discovery shook the world of astronomy and astrophysics with the announcement of the first extrasolar planets ever observed. These exoplanets were detected orbiting the millisecond pulsar PSR1257+12 (wer a immer si diese deppatn Namen ausdenkt die merkt si do ka Schwein). This discovery was made known through the letter titled "A Planetary System around the Millisecond Pulsar PSR1257+12" by Aleksander Wolszczan to Nature and a press release issued by Cornell University, which marked a pivotal moment in the exploration of exoplanets.
These newfound planets were identified through a technique called the pulsar timing method. By meticulously measuring the timing of radio signals emitted by the pulsar PSR1257+12, astronomers observed variations that were attributed to the gravitational influence of planets orbiting the pulsar. These observed pulsar period measurements over a few years were compared to the construction of a time model that included non-interacting Keplerian orbits and fitting it to the data, along with standard pulsar parameters [Figure 2](#pulsar). This approach allowed astronomers to deduce the properties of the detected planets and their orbits with a high degree of confidence (die confidence hättma a gern).

The discovery revealed two (or possibly three) planets with masses of 2.8 and 3.4 Earth masses, which was particularly intriguing because these planets were located in extremely close proximity to the pulsar. Their distances were about 0.47 and 0.34 AU. The most extraordinary aspect of this discovery lay in the unconventional environment where these planets were located, it was widely believed that planets could not exist around such exotic and hostile objects (aba sie san special deshalb können sies trotzdem).  The discovery prompted a reevaluation of our understanding of planetary formation and the remarkable diversity of planetary systems that might exist in the cosmos and raised questions about the diversity of planetary systems in the universe.
In summary, the 1992 discovery of exoplanets orbiting PSR1257+12 was a transformative event in exoplanetary science. It marked the first confirmed detection of planets beyond our solar system and challenged our understanding of planetary formation and the diversity of planetary systems in the universe (olles Grundwissn). This pivotal moment set the stage for continued investigation, offering the promise of deeper insights into these exoplanets and the extraordinary environments in which they exist. [(A. ## 2.4 Probability and cumulative distribution function

```python
plt.figure(figsize=(12, 8))
plt.suptitle('PDF and CDF using seaborn', fontsize = 16)

plt.subplot(2,2,1)
sns.histplot(data['mass'], kde=True)
plt.title('PDF for Mass')
plt.xlabel('Mass')

plt.subplot(2,2,2)
sns.ecdfplot(data['mass'])
plt.title('CDF for Mass')
plt.xlabel('Mass')

plt.subplot(2,2,3)
sns.histplot(data['radius'], kde=True)
plt.title('PDF for Radius')
plt.xlabel('Radius')

plt.subplot(2,2,4)
sns.ecdfplot(data['radius'])
plt.title('CDF for Radius')
plt.xlabel('Radius')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.suptitle('PDF and CDF using histograms and formulae', fontsize = 16)

plt.subplot(2,2,1)
plt.title("PDF for Mass")
N, bins_edges = np.histogram(mass, bins=30)
pdf_m = plt.plot(bins_edges[1:], N/sum(N))

plt.subplot(2,2,2)
plt.title("CDF for Mass")
cdf_m = plt.plot(bins_edges[1:], np.cumsum(N)/sum(N))

plt.subplot(2,2,3)
plt.title("PDF for Radius")
N, bins_edges = np.histogram(radius, bins=30)
pdf_r = plt.plot(bins_edges[1:], N/sum(N))

plt.subplot(2,2,4)
plt.title("CDF for Radius")
cdf_r = plt.plot(bins_edges[1:], np.cumsum(N)/sum(N))

plt.tight_layout()
plt.show()
```

These histograms should show the distribution of the amount of exoplanets discovered in specific planet groups, using restriction parameters. 

The catch while looking at these histograms, like discussed earlier, is again the difference in the scaling of the axes, so we have to be careful while interpreting and drawing assumptions (at this point afoch kane assumptions mehr machn). 

Probably the biggest difference between two plots is while comparing the exoplanets having a radius smaller than 0.15 jupyter units and those with a radius larger than 1 jupiter unit.

As we can see, way more exoplanets with radius larger than 1 jupyter unit have been discovered yet, which can be explained by the difficulty in detecting very small objects. The larger an exoplanet, the easier it is to be detected since it has more impact on it’s host star’s system, e.g. through it’s large gravitational impact. Big exoplanets potentially lead to a wobble of the star and, for our measured data, an asymmetry in the star’s light signal. The bigger the exoplanet, the bigger the impact on the host star and the bigger the asymmetry in our detected data is. 
The easiest and most common way is the radial velocity method which detects shifts in the stars spectrum. Now you might be thinking "why are these morons talking about spectrums, i thought we were looking at velocities", but we are in fact not wrong this time! <br>
The stars wobble -the velocity thingy caused by the planets pulling on the star- causes the star to seemingly accelerate towards earth and slow down again in a periodical time span. During these changes in velocity the light arriving at earth is either red or blue shifted (remember there was this dude with the Doppler effect). ## 4.3 Bootstrapping and regression

```python
x= np.log10(filtered_data_1['radius']) 
y= np.log10(filtered_data_1['mass'])
model = sm.OLS(y, sm.add_constant(x))
results= model.fit()

# This code utilizes the statsmodels library in Python for Ordinary Least Squares (OLS) linear regression.
#sm.add_constant(x) adding an intercept term to the independent variable 
#(Intercept/y-axis offset ist eine Konstante. Die allgemeine Form eines einfachen linearen Modells lautet: y=mx+b 
#wobei y die abhängige Variable,x die unabhängige Variable, m ist die Steigung (slope) der Linie, und b ist der Intercept. 
#Die Anwesenheit des Intercept-Terms in einem Regressionsmodell ermöglicht es der Regressionslinie, sich in der y-Richtung zu verschieben 
#und nicht notwendigerweise durch den Ursprung (0,0) zu gehen.)

#OLS minimizes the sum of squared differences between the actual and predicted logarithmic mass values, 
#identifying the line that best fits thelog-transformed data. 

num_bootstrap_samples = 200
bootstrap_slopes = []

for _ in range(num_bootstrap_samples):
    indices = np.random.choice(len(x), len(x), replace=True)
    x_bootstrap, y_bootstrap = x.iloc[indices], y.iloc[indices]
    results_bootstrap = sm.OLS(y_bootstrap, sm.add_constant(x_bootstrap)).fit()
    bootstrap_slopes.append(results_bootstrap.params.iloc[1])  # Updated line

# 1-sigma confidence interval for the slope
confidence_interval = np.percentile(bootstrap_slopes, [15.85, 84.15])

covariance = np.cov(x, y)[0, 1]
numeric_data = filtered_data_1.select_dtypes(include=[np.number])
corr_coef = numeric_data.corr().iloc[0, 1]

print(results.summary())
print(f"Slope: {results.params.iloc[1]:.3f} +/- {results.bse.iloc[1]:.3f}")  
print(f"Intercept: {results.params.iloc[0]:.3f} +/- {results.bse.iloc[0]:.3f}")  
print(f"Pearson Correlation Coefficient: {corr_coef:.3f}")
print(f"1-sigma Confidence Interval for the Slope: {confidence_interval}")
print('covariance', covariance)

plt.scatter(x, y, label='Data')
plt.plot(x, results.params.iloc[0] + results.params.iloc[1] * x, color='black', label='Linear Regression')  
plt.fill_between(x, results.params.iloc[0] + confidence_interval[0] * x, results.params.iloc[0] + confidence_interval[1] * x, color='red', label='Confidence Interval')
plt.xlabel('log(Radius)')
plt.ylabel('log(Mass)')
plt.legend()
plt.show()



x= np.log10(filtered_data['radius']) 
y= np.log10(filtered_data['mass'])
model = sm.OLS(y, sm.add_constant(x))
results= model.fit()

num_bootstrap_samples = 200
bootstrap_slopes = []

for _ in range(num_bootstrap_samples):
    indices = np.random.choice(len(x), len(x), replace=True)
    x_bootstrap, y_bootstrap = x.iloc[indices], y.iloc[indices]
    results_bootstrap = sm.OLS(y_bootstrap, sm.add_constant(x_bootstrap)).fit()
    bootstrap_slopes.append(results_bootstrap.params.iloc[1])  # Updated line

# 1-sigma confidence interval for the slope
confidence_interval = np.percentile(bootstrap_slopes, [15.85, 84.15])

covariance = np.cov(x, y)[0, 1]
numeric_data = filtered_data.select_dtypes(include=[np.number])
corr_coef = numeric_data.corr().iloc[0, 1]

print(results.summary())
print(f"Slope: {results.params.iloc[1]:.3f} +/- {results.bse.iloc[1]:.3f}")  
print(f"Intercept: {results.params.iloc[0]:.3f} +/- {results.bse.iloc[0]:.3f}")  
print(f"Pearson Correlation Coefficient: {corr_coef:.3f}")
print(f"1-sigma Confidence Interval for the Slope: {confidence_interval}")
print('covariance', covariance)

plt.scatter(x, y, label='Data')
plt.plot(x, results.params.iloc[0] + results.params.iloc[1] * x, color='black', label='Linear Regression')  
plt.fill_between(x, results.params.iloc[0] + confidence_interval[0] * x, results.params.iloc[0] + confidence_interval[1] * x, color='red', label='Confidence Interval')
plt.xlabel('log(Radius)')
plt.ylabel('log(Mass)')
plt.legend()
plt.show()
```

For the last plot, we used the bootstrap method. 

This method can be explained like the following (honestly kan Plan ob des exact richtig ist aba wir ham des so verstandn): if we e.g. have a sample of 1 000 variables, we pick a random (!) sub-set of let’s say 200 of those variables with replacement. We calculate let’s say the mean of our random sub-set. Then we repeat the processes of picking and calculation again and again and… you get our point here, we repeat it a lot. The means of the sub-sets can be compared to an estimation representing the whole sample. Basically we look at the mean of the means to find the best estimation (confused yet? No worries, we are too). What we’re trying to show is how accurate or “close” the estimated can get to the mean of the whole sample. This method is kinda like a play-around-and-see-what-happens game. Since with each trial, we get a different set of different randomly chosen variables, the mean of this sub-set changes with each trial. After repeating the process several times with different sub-sets, one might or hopefully get remarkably close to the actual mean of the whole sample. The bootstrap-slope gives the uncertainty of the estimations about the mean of the whole sample by using the sub-set mean (wir ham unser bestes gebn). 

After running the code several times, we finally got an estimated mean remarkably close to the actual mean of the whole sample space. The actual mean is 1.5343 and the estimated mean is 1.5352 (slay we did that).
